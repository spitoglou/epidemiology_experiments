```python
import numpy as np
import pandas as pd
import random
import pprint
import matplotlib.pyplot as plt

import math
from modelclass import Model
```

```python
# population
n=10000

#grid
x=100
y=100

# no of random contacts in one iteration
contacts=3

#probability of a contact to become infection
inf_prob = 0.3

# time to recovery
ttr = 5

# initial infected population
infected=2 #no

parameters = {
    'population': n,
    'grid': {
        'x': x,
        'y': y
    },
    'contacts': {
        'number': contacts,
        'type': 'random'
    },
    'initial_infected': infected,
    'probability_of_infection': inf_prob,
    'time_to_recovery': ttr
}

run = Model(parameters)
run.add_infected()
# print(run.stats)
# print(run.progression_dict)
# run.plot1()
run.plot2()

iterations = 15
for i in range(iterations):
    run.iterate()
    # pprint.pprint(run.progression_dict)
    # run.plot1()
    run.plot2()
run.plot_progression()
```
